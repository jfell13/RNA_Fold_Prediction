```python
print(len(in_df['fasta_id'].unique()))
def df_to_graph(in_df, distance_threshold=6.0):
    graphs = []
    for seq_id, group in in_df.groupby("fasta_id"):
        sequence = ''.join(group["resname"].tolist())  # e.g., 'AUGC'
        L = len(sequence)
        base_map = {'A': [1, 0, 0, 0], 'C': [0, 1, 0, 0], 'G': [0, 0, 1, 0], 'U': [0, 0, 0, 1]}
        x = torch.tensor([base_map[b] for b in sequence], dtype=torch.float32) # shape [L, 4]
        coords = group[["x_1", "y_1", "z_1"]].to_numpy() # "x" shape [L, 3]
        mean = coords.mean(axis=0)
        std = coords.std(axis=0) + 1e-8
        coords_norm = np.array((coords - mean) / std)
        y = torch.tensor(coords_norm, dtype=torch.float)
        if torch.isnan(y).any() == True:
            continue
        # Generate Distance matrix 
        diff_tensor = np.expand_dims(coords_norm, axis=1) - np.expand_dims(coords_norm, axis=0)
        distance_matrix = np.sqrt(np.sum(np.power(diff_tensor, 2), axis=-1))
        # Generate adjacency matrix "edge_index"
        # adj = distance_matrix < distance_threshold
        adj = (distance_matrix < distance_threshold) & (np.eye(L) == 0)
        u, v = np.nonzero(adj)
        edge_index = torch.tensor(np.array([np.concatenate([u, v]), np.concatenate([v, u])]), dtype=torch.long)
        # edge_index = torch.tensor([u, v], dtype=torch.long)
        # Add node edges based on adjacency matrix
        edge_attr = torch.tensor([[distance_matrix[start, end]] for start, end in zip(u, v)], dtype=torch.float)
        graph = Data(x=x, edge_index=edge_index, edge_attr=edge_attr, y=y)
        graphs.append(graph)
        
    return graphs

graphs = df_to_graph(in_df)
print(len(graphs))
```

```python
loader = DataLoader(graphs, batch_size=16, shuffle=True)
for batch in loader:
    print("Has NaNs in x:", torch.isnan(batch.x).any())
    print("Has NaNs in y:", torch.isnan(batch.y).any())
    break

for batch in loader:
    deg = degree(batch.edge_index[0], num_nodes=batch.num_nodes)
    print("Any isolated nodes?", (deg == 0).any().item())
    break
```

```python
class GCNNodeLevel(torch.nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels):
        super(GCNNodeLevel, self).__init__()
        self.conv1 = GCNConv(in_channels, hidden_channels, add_self_loops=True)
        self.conv2 = GCNConv(hidden_channels, hidden_channels, add_self_loops=True)
        # self.conv3 = GCNConv(hidden_channels, hidden_channels, add_self_loops=True)
        self.lin = torch.nn.Linear(hidden_channels, out_channels)

    def forward(self, data):
        x, edge_index, batch = data.x, data.edge_index, data.batch

        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = self.conv2(x, edge_index)
        # x = F.relu(x)
        # x = self.conv3(x, edge_index)
        # x = global_max_pool(x, batch)  # Need to pool node features to graph level for graph level tasks

        out = self.lin(x)

        return out
```

```python
def RMSD(pred, target):
    return torch.sqrt(((pred - target) ** 2).sum(dim=-1).mean())

def kabsch(P, Q):
    # P, Q: [L, 3] arrays
    C = np.dot(P.T, Q)
    V, S, W = np.linalg.svd(C)
    d = (np.linalg.det(V) * np.linalg.det(W)) < 0.0
    if d:
        V[:, -1] = -V[:, -1]
    U = np.dot(V, W)
    return U

def rmsd_kabsch(pred, true):
    P = pred - pred.mean(axis=0)
    Q = true - true.mean(axis=0)
    U = kabsch(P, Q)
    P_aligned = np.dot(P, U)
    return np.sqrt(np.mean((P_aligned - Q)**2))

def lddt_torch(predicted_points: torch.Tensor,
               true_points: torch.Tensor,
               cutoff: float = 15.0,
               per_residue: bool = False) -> torch.Tensor:
    """
    Computes approximate lDDT score between predicted and true 3D coordinates.

    Args:
        predicted_points: (L, 3) predicted 3D coordinates
        true_points: (L, 3) ground-truth 3D coordinates
        true_mask: (L,) binary tensor indicating valid positions
        cutoff: float, max distance in true structure to consider pair
        per_residue: bool, whether to return per-residue scores

    Returns:
        lDDT score (scalar) or (L,) tensor if per_residue=True
    """
    # Pairwise distances
    dmat_true = torch.cdist(true_points, true_points, p=2)
    dmat_pred = torch.cdist(predicted_points, predicted_points, p=2)

    # Mask for distances within cutoff in the *true* structure
    pair_mask = ((dmat_true < cutoff).float())

    # Absolute distance differences
    dist_l1 = torch.abs(dmat_true - dmat_pred)

    # Scoring: 4 thresholds
    score = 0.25 * ((dist_l1 < 0.5).float() +
                    (dist_l1 < 1.0).float() +
                    (dist_l1 < 2.0).float() +
                    (dist_l1 < 4.0).float())

    masked_score = pair_mask * score

    # Normalize
    if per_residue:
        norm = (pair_mask.sum(dim=1) + 1e-10)  # (L,)
        per_res_score = masked_score.sum(dim=1) / norm
        return per_res_score
    else:
        total_score = masked_score.sum()
        norm = pair_mask.sum() + 1e-10
        return total_score / norm

def coords_to_pdb(coords, chain_id='A', residue_name='A', atom_name="C1'", start_res=1):
    """
    coords: numpy array of coords
    Returns: string with PDB lines
    """
    pdb_lines = []
    atom_num = 1
    res_num = start_res

    for i in range(coords.shape[0]):
        
        x, y, z = coords[i]
        line = (
            f"ATOM  {atom_num:5d} {atom_name:<4} {residue_name:>3} {chain_id}{res_num:4d}    "
            f"{x:8.3f}{y:8.3f}{z:8.3f}  1.00  0.00           C"
        )
        pdb_lines.append(line)
        atom_num += 1
        res_num += 1

    pdb_lines.append("END")
    return "\n".join(pdb_lines)
```

```python
epochs = 50
# Initialize model, loss, and optimizer
model = GCNNodeLevel(in_channels=4, hidden_channels=64, out_channels=3)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
# loss_fn = torch.nn.MSELoss()
loss_fn = torch.nn.SmoothL1Loss()

loss_history = []
rmsd_history = []
kabsch_history = []
lddt_history = []

for epoch in range(epochs):
    model.train()
    total_train_loss = 0
    for batch in loader:
        optimizer.zero_grad()
        out = model(batch)
        loss = loss_fn(out, batch.y)
        loss_history.append(loss.mean().item())
        rmsd_train = RMSD(out, batch.y).mean().item()
        rmsd_history.append(rmsd_train)
        # kabsch_rmsd = rmsd_kabsch(out.detach().numpy(),batch.y.detach().numpy())#.mean().item()
        # kabsch_history.append(kabsch_rmsd)
        lddt = lddt_torch(out, batch.y)
        lddt_history.append(lddt)
        loss.backward()
        optimizer.step()
        total_train_loss += loss.item()
        
    print(f"Epoch {epoch}, Loss: {total_train_loss:.4f}, RMSD: {rmsd_train:.4f}")
    print(f"Kabsch RMSD: {kabsch_rmsd:.4f}, lDDT: {lddt:.4f}")

# Plot the training metrics after all epochs
# plot_metrics(log_file='train_metrics_log.txt')
```

```python
model.eval()
total_rmsd = 0
num_graphs = 0

with torch.no_grad():
    for batch in loader:
        pred = model(batch)

        # Assume batch is not batched (or handle batch slices manually)
        for i in range(batch.num_graphs):
            mask = batch.batch == i
            pred_i = pred[mask].cpu().numpy()
            true_i = batch.y[mask].cpu().numpy()
            break
            # Optional: denormalize if needed
            # pred_i = pred_i * batch[i].y.numpy().std() + batch.y[i].mean.numpy()
            # true_i = true_i * batch.y[i].std.numpy() + batch.y[i].mean.numpy()

            # RMSD with Kabsch
            rmsd_val = rmsd_kabsch(pred_i, true_i)
            total_rmsd += rmsd_val
            num_graphs += 1

print(f"Mean RMSD: {total_rmsd / num_graphs:.3f} Å")
```
